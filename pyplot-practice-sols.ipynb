### Construct a 2x2 plot with the following subplots
Using the Gaussian data below construct a figure with 2x2 subplots to meet the following specification. In each case annotate, title and label your plots, provide legends if appropriate, color etc. to improve readability and presentation.

1) Top left: plot a scatter plot of 1000 2d gaussian random variables, 500 drawn from a gaussian distribution with mean [0,0] and covariance [[1,0], [0,1]] and 500 from a gaussian distribution with mean [3,3] and covariance [[3,1], [1,1]].

2) Top right: plot sin(x) vs cos(x) for x between -pi and pi. Color them differently!

3) Simulate a fair coin flip 100 times (virtually using numpy), noting the number of times tails (or heads) is thrown. Do this 1000 times and plot a histogram of your results.

4) Bottem left: plot a heatmap of the function $f(x,y) = yx^2$ on $[-1,1]^2$

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Generate data for the plots

# Top left:
m = 100

mu1 = [0,0]
cov1 = [[1,0], [0,1]]

mu2 = [3,3]
cov2 = [[3,1], [1,1]]

x1,y1 = np.random.multivariate_normal(mu1, cov1, 500).T
x2,y2 = np.random.multivariate_normal(mu2, cov2, 500).T

# bottom left
coin_flips =np.random.binomial(100, 0.5, size=1000)

# bottom right
x = np.linspace(-1,1,50)
y = np.linspace(1,-1,50)
X,Y = np.meshgrid(x,y) 
Z = Y*X**2
```

```python
fig, ax = plt.subplots(3, 2,figsize = (10, 10))
ax[0,0].scatter(x1,y1,alpha=0.5, label = 'Gaussian 1')
ax[0,0].scatter(x2,y2,alpha=0.5, label = 'Gaussian 2')
ax[0,0].set(xlabel = "x", ylabel = "y", title = "Samples drawn from two Gaussians")

ax[0,1].plot(np.linspace(-np.pi, np.pi, 100), np.sin(np.linspace(-np.pi, np.pi, 100)), label='sin(x)')
ax[0,1].plot(np.linspace(-np.pi, np.pi, 100), np.cos(np.linspace(-np.pi, np.pi, 100)), label='sin(x)')
ax[0,1].set(xlabel = "x", title = "cos vs sin")
ax[0,1].legend()

ax[1,0].hist(coin_flips, alpha= .4)
ax[1,0].set(xlabel = "# simulations", ylabel = "# coin flips", title='Simulating sequences of coin flips')

ax[1,1].imshow(Z)
ax[1,1].set(xlabel = "x", ylabel = "y", title = "$f(x,y) = yx^2$")
divider = make_axes_locatable(ax[1,1])
cax = divider.append_axes('right', size='5%', pad=0.05)
im = ax[1,1].imshow(Z, extent=[-1,1,-1,1])
fig.colorbar(im, cax=cax, orientation='vertical')
```

Tips:
1) Always use fig, ax = plt.subplots(number rows , number cols,figsize = (10, 10))

2) Using indexing to access each different subplot, i.e., ax[r,c]

3) Populate a subplot or axis using a method, e.g., .plot(), .scatter(), .hist(). You can do this multiple times to create overlays. If you do this the alpha parameter for transparency becomes useful.
