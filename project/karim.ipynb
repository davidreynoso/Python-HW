```python
artwork = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-01-12/artwork.csv')
artists = pd.read_csv("https://github.com/tategallery/collection/raw/master/artist_data.csv")

artwork.to_csv("artwork.csv", index = False)
artists.to_csv("artists.csv", index = False)
artwork = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-01-12/artwork.csv')
artists = pd.read_csv("https://github.com/tategallery/collection/raw/master/artist_data.csv")

artwork["id"] = artwork["artistId"]
artwork = artwork[["id", "year", "acquisitionYear", "title", "medium"]]
artists = artists[["id", "gender"]]
df = pd.merge(artwork, artists)

def dimension(med_string):
    """
    Assign a dimension to a given piece of artwork based on the description
    of the medium, supplied as a string. 
    Media that include the words "paper", "canvas", "oil", or "paint" are assumed 
    2D. 
    Media that are not 2d and include the words "bronze", "stone", or "ceramic" are 
    assumed 3D. 
    Otherwise, the media is "Other/Unknown"
    
    @param med_string: str, the original medium
    @return dim: one of "2D", "3D", or "Other/Unknown" according to the rules above. 
    """
    if type(med_string) != str:
        med_string = str(med_string)
    med_string = med_string.lower()
    if any([w in med_string for w in ["paper", "canvas", "oil", "paint"]]):
        return "2D"
    elif any([w in med_string for w in ["bronze", "stone", "ceramic"]]):
        return "3D"
    else:
        return "Other/Unknown"

df["dimension"] = [dimension(m) for m in df["medium"]]
df = df[["title","acquisitionYear", "gender", "dimension"]]
df.head()
```

```python
# Get a DataFrame consisting of all the rows of acquisitionYear and gender
# for which the dimension is 2D

paintings = df.loc[df['dimension'] == '2D',['acquisitionYear','gender']]
display(paintings)

droppings = df[df['dimension'] == '2D'][['acquisitionYear','gender']]
display(droppings)
```

```python
# Use groupby to compute the number of male and female paintings over the last 10 years
tempdf = paintings[paintings['acquisitionYear'] >= 2013]  # when we subscript a dataframe by a series of true or false... 
                                                          # it returns the rows of the positions that are True
display(tempdf)
```

```python
# Replace bad values with NaN
decadePaintings = tempdf.dropna(axis=0) # axis 1 will remove the column, axis 0 will remove the row. matrices are always row by column!
display(decadePaintings)

newdf=decadePaintings.sort_values(['acquisitionYear']) # pointless as dataset only includes up to 2013
display(newdf)
```

- Describe what happens in k-fold cross validation. Why do we use k-fold cross validation as opposed to setting aside a fixed part of the training set for validation?

Now, in k-fold cross validation, we take the training set and split it up into k folds. We say one fold is a validation set, and the rest are used to train. we then score for this combination of train and validation, then move on to making the next fold our validation set. aggregate and mean to get a usable cross validation score.

- What is overfitting? How would you recognize it? What steps would you take to solve overfitting? Use polynomial regression as an example

Overfitting occurs when we have trained a model that has too high a complexity for the data it is trying to generalize. We can see this on a decision region as regions being really tight around certain datapoints, or when groups become non-continuous. We should reduce the complexity. In the case of polynomial regression, we should reduce the degree of the polynomial function we are using to estimate our predictor variable. Overfitted models will not generalize well. Overfit models may score very well on train data but poorly on test data.

- What is underfitting? How would you recognize it? What steps would you take to solve underfitting? Use a decision tree as an example.

Underfitting occurs as a result of not having enough complexity in your model. We can recognize underfit models with decision regions and scores. Underfit models will typically not generalize well because they did not have enough complexity to fit the training data. Underfit models will score low on both train and test.

```python
url = "https://philchodrow.github.io/PIC16A/datasets/titanic.csv"
titanic = pd.read_csv(url)
le = preprocessing.LabelEncoder()
titanic['Sex'] = le.fit_transform(titanic['Sex'])
titanic.head()
```

```python
train, test = train_test_split(titanic[["Pclass", "Age", "Sex", "Survived"]], test_size = 0.3)

X_train = train[["Pclass", "Age", "Sex"]]
y_train = train["Survived"]
X_test = test[["Pclass", "Age", "Sex"]]
y_test = test["Survived"]
```

```python
# Fit a tree of depth 2 to the data to predict survivability based on 
# sex, age and Pclass and outputs its training and test score
n = 2
myTree = tree.DecisionTreeClassifier(max_depth=n)

myTree.fit(X_train,y_train)

print("The DecisionTreeClassifier with max_depth param =",n,"scored:\n",myTree.score(X_train,y_train),"on train data.\n",myTree.score(X_test,y_test),"on test data.")
```

```python
# Fit a tree of depth 8 to the data to predict survivability based on 
# sex, age and Pclass and outputs its training and test score
n = 8
myTree = tree.DecisionTreeClassifier(max_depth=n)

myTree.fit(X_train,y_train)

print("The DecisionTreeClassifier with max_depth param =",n,"scored:\n",myTree.score(X_train,y_train),"on train data.\n",myTree.score(X_test,y_test),"on test data.")
```

```python
# Fit a tree of depth100 to the data to predict survivability based on 
# sex, age and Pclass and outputs its training and test score
n=100
myTree = tree.DecisionTreeClassifier(max_depth=n)

myTree.fit(X_train,y_train)

print("The DecisionTreeClassifier with max_depth param =",n,"scored:\n",myTree.score(X_train,y_train),"on train data.\n",myTree.score(X_test,y_test),"on test data.")
```

```python
train, test = train_test_split(titanic[["Pclass", "Age", "Sex", "Fare"]], test_size = 0.3)

X_train = train[["Fare", "Age", "Sex"]]
y_train = train["Pclass"]
X_test = test[["Fare", "Age", "Sex"]]
y_test = test["Pclass"]
```

```python
# Now use cross validation to train an SVC to predict passenger class using 
# passenger age, sex and fare. Ue cross validation to find a good choice of gamma.
# In particular, plot a graph of cv score vs gamma with a logarithmic x axis.
from sklearn import svm
from sklearn.model_selection import cross_val_score

bestScore = 0
bestGamma = 0.01
gammas = np.arange(0.01,1.00,0.01)
cvs=[]
for i in gammas:
    vec = svm.SVC(gamma=i)
    cvs.append(cross_val_score(vec,X_train,y_train,cv=5).mean())
    if cvs[-1] > bestScore:
        bestScore = cvs[-1]
        bestGamma = i
plt.plot(gammas,cvs)
plt.xscale('log')
```

```python
bestScore,bestGamma
```

```python
vectorMachine = svm.SVC(gamma=bestGamma)
vectorMachine.fit(X_train,y_train)
print("vectorMachine scores with gamma parameter of", bestGamma,'\n','Train score:',vectorMachine.score(X_train,y_train),'\n','Test score:',vectorMachine.score(X_test,y_test))
```

### Algorithmic bias

- Define in your own words sample bias, population (or historical) bias and measurement bias.

Sample bias means that the data we are trying to train our model with does not actually generalize for our needs

Population or historical bias means that dominant from historical time periods may not necessarily fit with the trends of today. the model would be a poor predictor of current data, as it is biased by the historical data.

Measurement bias is the error that exists in every measurement we make. There is some degree of imprecision at any level. Or.. the target variables do not actually allow us to abstract any useful information
