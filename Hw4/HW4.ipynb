## Problem 1: Faceted Histogram

Run the following code block to define a function which generates two 1-dimensional `numpy` arrays. The first array, called `groups`, consists of integers between `0` and `n_groups - 1`, inclusive. The second array, called `data`, consists of real numbers. 

````python
import numpy as np
from matplotlib import pyplot as plt

def create_data(n, n_groups):
    """
    generate a set of fake data with group labels. 
    n data points and group labels are generated. 
    n_groups controls the number of distinct groups. 
    Returns an np.array() of integer group labels and an 
    np.array() of float data. 
    """
    
    # random group assignments as integers between 0 and n_groups-1, inclusive
    groups = np.random.randint(0, n_groups, n)
    
    # function of the groups plus gaussian noise (bell curve)
    data   = np.sin(groups) + np.random.randn(n)
    
    return(groups, data)
````

````python
groups, data = create_data(1000, 6)
````

### Requirements: 

1. Your function should work **whenever `m_rows*m_cols` is equal to the total number of groups.** Your function should first check that this is the case, and raise an informative `ValueError` if not. You may assume that there is at least one data point for each group label in the data supplied. 
2. For full credit, you should not loop over the individual entries of `groups` or `data`. It is acceptable to loop over the distinct values of `groups`. In general, aim to minimize `for`-loops and maximize use of `Numpy` indexing.

### Hints

- If your plots look "squished," then `plt.tight_layout()` is sometimes helpful. Just call it after constructing your figure, with no arguments. 
- Integer division `i // j` and remainders `i % j` are helpful here, although other solutions are also possible. 


````python
# your solution here
def facet_hist(groups,data,m_r,m_c,figsize, **kwargs):
    """Creates a faceted histogram of different labeled groups, with the corresponding data to each elt. 
    Args:
        groups: np.array of group labels as output by create_data()
        data: np.array of data as output by create_data()
        m_r: the number of desired rows in your faceted histogram
        m_c: the number of desired columns in your faceted histogram
        figsize: the size of the figure
        **kwargs: kwargs that will customize the histograms appearance
    Returns:
        a faceted histogram abiding by the parameters structure
    """
    # we check that our rows and collums can run with the amount of groups, otherwise valueerror
    if (m_r*m_c) != (np.max(groups) + 1):
        raise ValueError("we need m_rows*m_cols equal to the total number of groups.")
    fig, ax = plt.subplots(m_r, m_c,figsize = figsize)
    # loop through each group and create a histogram by using integer division to decide the rows,
    # mods to decide the collumns, where we take the indexs of the groups and match it to its data point
    # we then specify where we want our labels
    for i in range(m_r*m_c): 
        t = "Group " + str(i)
        r = i//m_c
        ax[r,i%m_c].hist(data[np.where(groups == i)[0]], **kwargs)
        ax[r,i%m_c].set(title = t)
        if r == m_r -1:
            ax[r,i%m_c].set(xlabel = "x")
        if (i%m_c) == 0:
            ax[r,i%m_c].set(ylabel = "Frequency")
    plt.tight_layout()
````

Write a function called `scatterplot_matrix()`, with arguments `cols` and `figsize`. The `cols` argument should be a list of strings, each of which are the name of one of the columns above, for example `cols = ["Score", "GDP per capita", "Social support"]`. There is a separate scatterplot for each possible pair of variables. In fact, there are two: one where the first variable is on the horizontal axis, and one where it's on the vertical axis. Some analysts prefer to remove half the plots to avoid redundancy, but you don't have to bother with that. The diagonal is empty, since there's no point in investigating the relationship between a variable and itself. 

Don't forget comments and docstrings! 

````python
# define your function
def scatterplot_matrix(cols, figsize):
    """Create faceted scatterplot of each relationship is a given list with the panda table 
    Args:
        cols: list of a strings - using names of the collumns in our 'happiness' table
        figsize: the size of the figure
    Returns:
        a faceted scatterplot abiding by the parameters structure
    """
    fig, ax = plt.subplots(len(cols), len(cols),figsize = figsize)
    # use a nested loop to access each scatterplot w/ indexing
    for i in range(len(cols)):
        for j in range(len(cols)):
            # label the plots using the nested loop
            ax[i,j].set(ylabel = cols[j], title = cols[i])
            # ignore the plots of two same labels
            if cols[i] != cols[j]:
                # take the correlation coeff of current comparison, create a scatter, then label the coeff
                my_number = np.corrcoef(happiness[cols[i]],happiness[cols[j]])
                ax[i,j].scatter(happiness[cols[i]],happiness[cols[j]])
                ax[i,j].set(xlabel = r"$\rho$ = " + str(np.round(my_number[0,1], 2)))
        plt.tight_layout()
````

Run the following code to download two time series data sets:

## Part A

The series `COVID['cases']` is essentially a `numpy` array containing the cumulative case counts over time. The COVID19 case data  is cumulative, but we would like to see the number of new cases per day (i.e. as in [this kind of plot](https://www.google.com/search?q=covid+stats)). Check the documentation for the `np.diff` function and figure out what it does. Use it appropriately to construct a new array, called `per_day`, giving the number of new cases per day. Then, make a new array called `per_day_date` that gives the appropriate date for each case count. In particular, you will need to ensure that `per_day` and `per_day_date` have the same shape. 

````python
# your solution here
# use np.diff on covid cases to see the diffence from each previous day to present
per_day = np.diff(COVID['cases'])
# create a temporary numpy array of the dates
temp = np.array(COVID['date'])
# notice our per_day has one less value as the first day doesnt have a previous day to have a diff,
# so we cut off our first date to match shape, reshape to that of per_day, and notice the dates now match the 
# new number of cases per day
per_day_date = temp[1:].reshape(1128,)
````
